```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data2.csv", index_col=0)
df = df.transpose()
df.head()
```

```python
plt.plot(df, marker="o")
plt.legend(df.columns.values)
plt.xlabel("Number of Users")
plt.ylabel("Throughput (# Users / Second)")
plt.grid()
plt.title("Throughput vs Number of Users")
plt.show()
```
